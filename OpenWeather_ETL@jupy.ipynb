```python
import json
import pandas as pd
import requests
```

```python
api_key='#'
w_url='https://api.openweathermap.org/data/2.5/weather?'
```

```python
w_data_list=[]
city=['Washington','Chicago','Dallas','Atlanta','San Francisco','Los Angeles','Houston','San Diego','San Antonio','Austin','Philadelphia',
      'New York','London','Visakhapatnam','Bengaluru','Hyderabad','Chennai','Delhi','Mumbai','Pune','Dubai','Riyadh','Abu Dhabi','Doha','Muscat',
      'Gaza','Beirut','Tehran','Jerusalem','Manama','Baghdad','Kabul','Lahore','Karachi','Islamabad','Dhaka','Tokyo','Seoul','Beijing','Shanghai',
      'Taipei','Busan','Cairo','Lagos','Nairobi']
for c in city:
    params={
        'q':c,
        'appid':api_key,
        'units':'metric'
    }
    w_res=requests.get(w_url,params=params)
    if w_res.status_code==200:
        w_data=w_res.json()
        w_data_list.append(w_data)
print(w_data_list)
    
```

```python
w_df.rename(columns={'max_temp':'max_temp C','min_temp':'min_temp C','pressure':'pressure hPa','sea_level':'sea_level hPa','humidity':'humidity %',
                     'sunrise':'sunrise UTC','sunset':'sunset UTC','timezone':'timezone UTC'
                    },inplace=True)
w_df
```

```python
from datetime import datetime
convert_rise=[]
for i in w_df['sunrise UTC']:
    dt_obj=datetime.fromtimestamp(i)
    sunrise_time=dt_obj.strftime('%Y-%m-%d %H:%M:%S')
    convert_rise.append(sunrise_time)
w_df['sunrise UTC']=convert_rise
convert_set=[]
for i in w_df['sunset UTC']:
    dt_obj=datetime.fromtimestamp(i)
    sunset_time=dt_obj.strftime('%Y-%m-%d %H:%M:%S')
    convert_set.append(sunset_time)
w_df['sunset UTC']=convert_set
w_df
```

```python
timezone_list=[]
for i in w_df['timezone UTC']:
    k=i/3600
    if k>0:
        c='UTC+'+str(int(k))
        timezone_list.append(c)
    else:
        c='UTC'+str(int(k))
        timezone_list.append(c)
w_df['timezone UTC']=timezone_list
w_df
```
